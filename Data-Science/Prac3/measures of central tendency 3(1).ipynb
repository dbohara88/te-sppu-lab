### Measures of central tendency

```python
# Min
df.min()
```

```python
# Max
df.max()
```

```python
# Mode
df['age'].mode()
```

```python
# Mean
df['age'].mean()
```

```python
# Median
df['age'].median()
```

```python
# Std Deviation
round(df['age'].std(),2)
```

### Summary Statistics

```python
# Summary statistics for all numerical columns
round(df.describe(),2)
```
